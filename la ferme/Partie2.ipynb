```python
import pandas as pd
```

```python
pluie = pd.read_csv('pluie.csv')
temperature = pd.read_csv('temperature.csv',encoding = "ISO-8859-1",sep = ';')
temperature
```

```python
for i in range(len(temperature)):
    if temperature.moyenne[i] > temperature.maxi[i]:
        temp = temperature.moyenne[i]
        temperature.moyenne[i] = temperature.maxi[i]
        temperature.maxi[i] = temp
temperature
```

```python
for i in range(len(temperature)):
    if temperature.moyenne[i] == temperature.maxi[i]:
        temperature.mini = temperature.maxi[i]
```

```python
for i in range(len(pluie)):
    if pluie.terrain[i] == "soixante sept":
        pluie.terrain[i] = 67
    if pluie.terrain[i] == "dix sept":
        pluie.terrain[i] = 17
    if pluie.terrain[i] == "treize":
        pluie.terrain[i] = 13
    if pluie.terrain[i] == "treize":
        pluie.terrain[i] = 13
    if pluie.terrain[i] == "vingt cinq":
        pluie.terrain[i] = 25
    if pluie.terrain[i] == "trente trois":
        pluie.terrain[i] = 33
    if pluie.terrain[i] == "quarante sept":
        pluie.terrain[i] = 47
    if pluie.terrain[i] == "quinze":
        pluie.terrain[i] = 15
    if pluie.terrain[i] == "quarante quatre":
        pluie.terrain[i] = 44
    if pluie.terrain[i] == "quarante et un":
        pluie.terrain[i] = 41
    if pluie.terrain[i] == "quatre vingt":
        pluie.terrain[i] = 80
    if pluie.terrain[i] == "soixante neuf":
        pluie.terrain[i] = 69
pd.to_numeric(pluie.precipitation)
```

```python
pluie = pluie.groupby(['saison','annee','terrain']).sum().reset_index()
pluie['terrain'] = pluie['terrain'].astype('int64')
pluie
```

```python
pluie = pluie.groupby(['saison','annee','terrain']).sum().reset_index()
pluie
```

```python
countete = 0
countp = 0
counthiver = 0
countautomne = 0
tote = 0
toth = 0
tota = 0 
totp = 0
for i in range(len(pluie)):
    if pluie.precipitation[i] > 2300 and pluie.precipitation[i] < 5500:
        if pluie.saison[i] == "automne":
            pluie.precipitation[i] = pluie.precipitation[i] / 10
        if pluie.saison[i] == "ete":
            pluie.precipitation[i] = pluie.precipitation[i] / 100
    if pluie.precipitation[i] > 5500:
        if pluie.saison[i] == "ete":
            countete = countete + 1
            tote = pluie.precipitation[i] + tote
        if pluie.saison[i] == "automne":
            countautomne = countautomne + 1
            tota = pluie.precipitation[i] + tota
        if pluie.saison[i] == "hiver":
            counthiver = counthiver + 1
            toth = pluie.precipitation[i] + toth
        if pluie.saison[i] == "printemps":
            countp = countp + 1
            totp = pluie.precipitation[i] + totp
for i in range(len(pluie)):
    if pluie.precipitation[i] > 5500:
        if pluie.saison[i] == "automne":
            if countautomne == 1:
                pluie.precipitation[i] = 200
            if countautomne > 1:
                pluie.precitation = tota / countautomne 
        if pluie.saison[i] == "ete":
            if countete == 1:
                pluie.precipitation[i] = 200
            if countete > 1:
                pluie.precitation = tote / countete
        if pluie.saison[i] == "hiver":
            if counthiver == 1:
                pluie.precipitation[i] = 200
            if counthiver > 1:
                pluie.precitation = toth / counthiver
        if pluie.saison[i] == "printemps":
            if countp == 1:
                pluie.precipitation[i] = 200
            if countp > 1:
                pluie.precitation = totp / countp
```
